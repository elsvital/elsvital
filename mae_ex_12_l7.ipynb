$$ f_Y(u) = 0.16e^{(-0.16u)} \mathbb{1}_{(0,\infty)}(u) $$

\begin{eqnarray} 
P(Y>t) &=& 1-P(Y \leq t) \\
&=& 1 - \int_0^{t} f_Y(u)du\\
&=& 1 - \int_0^{t} 0.16e^{(-0.16u)}du\\
&=& 1 - (1-e^{-0.16t})\\
&=& e^{-0.16t}, t > 0 \in \mathbb{R}
\end{eqnarray}

```python
import matplotlib.pyplot as plt
import pandas as pd
import random
import numpy as np

x = []
xp = []
y = []
yp = []
z = []
zp = []


l = 0.16
n = 200

t = np.linspace(0, 200)
def fd(x,l):
    return 1-np.exp(-l*x)
#plt.subplot(3, 1, 3)
#plt.scatter(t,fd(t,l))

plt.figure();
fis = pd.Series(fd(t,l),t)
#fis = s.sort_values(ascending=True)
fis.plot()
```

```python
def fx_inv(u,l):
    return (-np.log(u)+np.log(l))/l

def fx(u,l):
    return l*np.exp(-l*u)

for i in range(n):
    u = random.uniform(0,1)
    x.append(u)
    y1= fx_inv(u,l)
    y.append(y1)
    z1 = fx(y1,l)
    if (y1 > 0):
        yp.append(y1)
        xp.append(u)
        zp.append(z1)
    z.append(z1)
    
plt.figure();
fis = pd.Series(xp,yp)
fis = fis.sort_values(ascending=True)
fis.plot()
plt.figure();
fs = pd.Series(y,z)
fs = fis.sort_values(ascending=True)
fs.plot()

df4 = pd.DataFrame({"obs_200": yp})
plt.figure()
df4.plot.hist(density=True)
fs.plot()
```
